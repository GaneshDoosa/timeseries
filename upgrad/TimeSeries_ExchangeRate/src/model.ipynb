```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
exchange_data = pd.read_csv('../dataset/exchange-rate-twi.csv')
exchange_data
```

```python
train_len = 212
train = exchange_data[:train_len]
test = exchange_data[train_len:]
y_hat = test.copy()
```

```python
# Naive method
y_hat['naive_forecast'] = train['Exchange Rate TWI'][train_len-1]
```

```python
# Simple average method
y_hat['avg_forecast'] = train['Exchange Rate TWI'].mean()
```

```python
y_hat
```

```python
# RMSE and MAPE
from sklearn.metrics import mean_squared_error
rmse = np.round(np.sqrt(mean_squared_error(y_hat['Exchange Rate TWI'], y_hat['naive_forecast'])),2)
mape = np.round(np.mean(np.abs(y_hat['Exchange Rate TWI']-y_hat['naive_forecast'])/y_hat['Exchange Rate TWI'])*100,2)
results = pd.DataFrame({'Method':['Naive Method'], 'RMSE':rmse, 'MAPE':mape, 'Accuracy': 100-mape})
results
```

```python
rmse = np.round(np.sqrt(mean_squared_error(y_hat['Exchange Rate TWI'],y_hat['avg_forecast'])),2)
mape = np.round(np.mean(np.abs(y_hat['Exchange Rate TWI']-y_hat['avg_forecast'])/y_hat['Exchange Rate TWI'])*100,2)
results = results.append({'Method': 'Simple Average', 'RMSE':rmse, 'MAPE':mape, 'Accuracy':100-mape}, ignore_index=True)
results
```
